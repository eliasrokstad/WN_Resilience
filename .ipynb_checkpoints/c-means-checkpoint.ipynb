```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
import seaborn as sns
import os

def absoluteFilePaths(directory):
    for root, dirs, files in os.walk(os.path.abspath(directory)):
        return [os.path.join(root, file) for file in files]



def get_resilience(directory):
    results_files = absoluteFilePaths(directory)
    df = pd.DataFrame()
    for file in results_files:
        temp_df = pd.read_csv(file, index_col=0).groupby(['Dimension', 'Aspect', 'Threshold']).mean().reset_index()
        temp_df = temp_df[['Dimension', 'Aspect', 'Threshold', 'Reliability']].rename(columns={'Reliability': 'Resilience'})
        temp_df['Network'] = file.split('_')[1]
        df = df.append(temp_df, ignore_index=True)
    return df
```

```python
#raw_data = get_resilience('./results')
raw_data = pd.read_csv('results/resilience_normal')
print(raw_data.head())
```

```python
c1 = (raw_data['Dimension'] == 'Magnitude')
c2 = (raw_data['Aspect'] == 'Service') 
c3 = (raw_data['Threshold'] == 0)
df = raw_data[c1 & c2 & c3].set_index('Network')
print(df['Resilience'].min()-df['Resilience'].max())
```

```python
def get_series(dim, aspect, threshold):
    c1 = (raw_data['Dimension'] == dim)
    c2 = (raw_data['Aspect'] == aspect) 
    c3 = (raw_data['Threshold'] == threshold)
    df = raw_data[c1 & c2 & c3].set_index('Network')
    return df['Resilience']

print(raw_data['Dimension'].unique())
series = {'S': get_series('Magnitude', 'Service', 0),
          r'$S_{peak}$': get_series('Peak', 'Service', 0),
          'N': get_series('Magnitude', 'Spatial', 0),
          r'$N_{peak}$': get_series('Peak', 'Spatial', 0)}
df = pd.DataFrame(series)
print(df.head())
```

```python
from itertools import combinations
columns = df.columns
scaler = MinMaxScaler()
for col in columns:
    df[col] = scaler.fit_transform(df[[col]])

graph_combinations = [comb for comb in combinations(columns, 2)]

```

```python
distortions = []
K = range(1,10)

for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(df[columns])
    distortions.append(kmeanModel.inertia_)
    

plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
km = KMeans(n_clusters=3)
y_predict = km.fit_predict(df[columns])
df['cluster'] = y_predict
```

```python
groups = df.groupby('cluster').mean().sum(axis=1)
for cluster, performance in zip(groups.sort_values().iteritems(), ['Low performance', 'Medium performance', 'High performance']):
    df.loc[df['cluster'] == cluster[0], 'cluster'] = performance


print(df.head())
```

```python
sns.set_context("paper", font_scale=2)


plt.rcParams['figure.figsize'] = [15, 10]
sns.set_style("white")
sns.set_palette("bright")
#palette = sns.color_palette("Spectral", as_cmap=True)
fig, axs = plt.subplots(2, 3)
i = 0
for x, ax_x in enumerate(axs[0, :]):
    for y, ax_y in enumerate(axs[:, 0]):
        if x == 1 and y == 0:
            sns.scatterplot(ax= axs[y, x], x=df[graph_combinations[i][0]], y= df[graph_combinations[i][1]], hue=df['cluster'], style=df['cluster'], s=15)
            sns.move_legend(axs[y, x], "lower center", bbox_to_anchor=(0.45, 1), ncol=4, title=None, frameon=False)
        else:
            sns.scatterplot(ax= axs[y, x], x=df[graph_combinations[i][0]], y= df[graph_combinations[i][1]], hue=df['cluster'], style=df['cluster'], s=15, legend = False)
        axs[y, x].set_xticks([])
        axs[y, x].set_yticks([])

        i += 1

        

plt.subplots_adjust(wspace=0.15, hspace=0.1)
plt.savefig('scatter.png', bbox_inches='tight')
plt.show()


```

```python
attributes_data = pd.read_csv('attributes.csv')
for n, row in attributes_data.iterrows():
    attributes_data.loc[n, 'Variant'] = row['Variant'].split('_')[-1]
attributes_data = attributes_data.rename(columns={"Variant": "Network"})
attributes_data = attributes_data[attributes_data['Network'].isin(df.index.to_list())]
attributes_data = attributes_data.set_index('Network')
attributes_columns = attributes_data.columns
attributes_data['cluster'] = df['cluster']
attributes_data['Group'] = df['cluster']

print(attributes_data['cluster'].unique())
print(df['cluster'].unique())
print(attributes_data.columns)
```

```python
sns.set_context("paper", font_scale=2)

plt.rcParams['figure.figsize'] = [15, 10]
sns.set_style("white")
sns.set_palette("bright")


attributes_data['cluster'] = pd.CategoricalIndex(attributes_data['cluster'], ordered=True, categories=df['cluster'].unique())
sns.set_style("white")
sns.set_palette("Spectral")
columns = ['Link density', 'Clustering Coefficient', 'Central Point of Dominance', 'Bridge density']
print(columns)
fig, axs = plt.subplots(len(columns), 1)

rename_att = {'Link density': 'd',
           'Average shortest path length': 'a',
            'Graph diameter': r'$C_c$',
           'Clustering Coefficient': r'$C_d$',
           'Graph radius': r'$G_d$',
           'Central Point of Dominance': r'$G_r$',
           'Bridge density': r'$B_d$'}

for x, ax_x in enumerate(axs[:]):
    sns.boxplot(data=attributes_data, x='cluster', y=columns[x], ax=axs[x], showfliers = True, palette='bright')
    axs[x].set_yticks([])
    #print(axs[x].get_ylabel())
    axs[x].set_ylabel(rename_att[axs[x].get_ylabel()])


plt.subplots_adjust(wspace=1, hspace=0)
plt.savefig('box_group.png', bbox_inches='tight')
plt.show()
```
