```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import os
```

```python
resilience = pd.read_csv('results/resilience_normal.csv', index_col=0)
attributes = pd.read_csv('results/attributes_normal.csv', index_col=0)
resilience = resilience.drop('benchmark')
```

```python
resilience.head()
```

```python
attributes.head()
```

```python
df = pd.concat([resilience, attributes], axis=1)
corr = df.corr()
plt.rcParams['figure.figsize'] = [10, 5]
sns.set_style("whitegrid")
sns.heatmap(corr, cmap="Spectral");
```

```python
corr = df.corr()
corr = corr.drop(attributes.columns.to_list(), axis=0)
corr = corr.drop(resilience.columns.to_list(), axis=1)

plt.rcParams['figure.figsize'] = [10, 5]
sns.set_style("whitegrid")
sns.heatmap(corr, cmap="Spectral");
```
